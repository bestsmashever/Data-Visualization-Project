```python
# import dependencies
import pandas as pd
import requests
from bs4 import BeautifulSoup
from splinter import Browser
from splinter.exceptions import ElementDoesNotExist
import re
import time
import os
import csv
```

```python
# set the chromedriver path
executable_path = {"executable_path": "chromedriver.exe"}
browser = Browser("chrome", **executable_path, headless=False)
```

```python
# url
job_title = "Data%20Scientist"
location_id = "1139761"
location = "Austin, %20TX"
glassdoor_url = "https://www.glassdoor.com/Job/jobs.htm?sc.keyword="+job_title+"&locT=C&locId="+location_id+"&locKeyword="+location+"&srs=RECENT_SEARCHES"
glassdoor_url
```

```python
browser.visit(glassdoor_url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# find the xpath and return a collection of all jobs on a certain page
xpath = "//*[@id='MainCol']/div/ul/li"
results = browser.find_by_xpath(xpath)
companies = []
ratings = []
salaries = []
job_descriptions = []
```

```python
for res in results:
    try:
        res.click()
        salaries.append(res.find_by_css("div:nth-child(3) > div:nth-child(1) > span").text)
        
        company = res.find_by_css("div.flexbox.empLoc > div").text
        companies.append(company.split('–')[0])
        
        ratings.append(res.find_by_css("div.logoWrap > span").text)
        job_descriptions.append(browser.find_by_id("JobDescriptionContainer").text)
    except:
        pass
		
companies = []
ratings = []
salaries = []
job_descriptions = []
xpath = "//*[@id='MainCol']/div/ul/li"
def save_results():
    results = browser.find_by_xpath(xpath)
    for res in results:
        try:
            if not (res.find_by_css("div:nth-child(3) > div:nth-child(1) > span").text):
                print("not found")
                break
            if not (res.find_by_css("div.flexbox.empLoc > div").text):
                print("not found")

                break
            if not (res.find_by_css("div.logoWrap > span").text):
                print("not found")
                break
            if not (browser.find_by_id("JobDescriptionContainer").text):
                print("not found")

                break
                
            salaries.append(res.find_by_css("div:nth-child(3) > div:nth-child(1) > span").text)
            company = res.find_by_css("div.flexbox.empLoc > div").text
            companies.append(company.split('–')[0])
            ratings.append(res.find_by_css("div.logoWrap > span").text)
            job_descriptions.append(browser.find_by_id("JobDescriptionContainer").text)
        except:
            continue

def close_modal():
    if len(browser.find_by_css(".xBtn")):
        browser.find_by_css(".xBtn")[0].click()

def is_last_page():
    if len(browser.find_by_css("li.page.current.last")):
        return True
    return False

def next_page():
    next_pages = browser.find_by_css("li.next")
    next_pages[0].click()



last_page = False
counter = 0
while not last_page:
    if is_last_page(): last_page = True
    close_modal()
    print(f"Processing Page {counter+1}")
    save_results()
    counter += 1
    next_page()
```
