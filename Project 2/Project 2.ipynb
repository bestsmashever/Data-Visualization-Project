```python
def get_job_info(soup):
    # get job titles
    jobs = []
    titles = soup.find_all('div', class_ = 'jobTitle')
    for title in titles:
        link = title.find('a', class_='jobLink')
        if link:
            jobs.append(link.text)
    jobs = list(filter(None,jobs))

    # get employer name and locations
    employers = soup.find_all('div', class_='flexbox empLoc')
    companies = []
    locations = []
    for employer in employers:
        emp_loc = employer.text
        split = emp_loc.split('–')
        companies.append(split[0].strip())
        locations.append(split[1].split('TX')[0] + 'TX')
        
        
    # get salary range (if possible)
    print(jobs)
    print(len(jobs))

    # create dictionary
    d = {
    'Job Name': jobs,
    'Company': companies,
    'Location': locations
    }
    
    # create dataframe
    job_search_df = pd.DataFrame(d)
    # return dataframe
    return job_search_df
```
